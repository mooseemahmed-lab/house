## step-9 Price Distribution Graph

```python
plt.hist(df["Price"])
plt.title("House Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

# step-8 Bedrooms vs Price Graph

```python
plt.scatter(df["Bedrooms"], df["Price"])
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.title("Bedrooms vs Price")
plt.show()
```
